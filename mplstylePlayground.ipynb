```python
%pylab inline
import matplotlib.pyplot as plt
import numpy as np
from pathlib import Path
```

## test output

```python
plt.style.use('haruiium-lightside.mplstyle')
plt.rcParams['text.usetex'] = True

expName = 'teste'

t = np.linspace(0.0, 1.0, 50)
s = np.cos(4 * np.pi * t) + 2

fig, ax = plt.subplots(figsize=(6, 4), tight_layout=True)
for i in range(10):
    ax.plot(t, s+i*1, marker='.', label=f"i={i}")

ax.grid(linestyle='--')
ax.set_xlabel(r'\textbf{time (s)}')
ax.set_ylabel("$\\log_{2}{\\frac{1}{n} \\ \\sum_{t=0}^{n} | [{Tr}({\\rho_A}^2)]_e - [{Tr}({\\rho_A}^2)]_t | }$", fontsize=16)
ax.set_title(r'\TeX\ is Number $\displaystyle\sum_{n=1}^\infty'
             r'\frac{-e^{i\pi}}{2^n}$!', fontsize=16)

legendPlt = ax.legend(
        bbox_to_anchor=(1.025, 1.0),
        loc='upper left',
        borderaxespad=0.)
demoPlotFig = plt.savefig(
        Path('./demo-lightside.png'),
        format='png',
        dpi=800,
        bbox_extra_artists=(legendPlt, ),
        bbox_inches='tight',
    )
```
